Here, we bring the building blocks togehter

```
from pipeline_elements.parser import parser
from pipeline_elements.solving import optimize
from pipeline_elements.submission_and_score import create_submission

import numpy as np
import time
```

### Run the pipeline

```
DESCRIPTION = "Optimizer1"

print("Start Pipeline!")
FILENAMES =["a_example.in", 
            "b_little_bit_of_everything.in",
            "c_many_ingredients.in",
            "d_many_pizzas.in",
            "e_many_teams.in"
            ]

scores = []
times = []

for name in FILENAMES:
    
    print(f"START PROBLEM {name}")

    print(f"-----Get the Data for problem {name}")
    lookup = parser(name)

    tic = time.time()
    print("-----Optimize the Problem ... ")
    result, score = optimize(lookup)

    toc = time.time()
    
    scores.append(score)
    times.append(toc-tic)

    print("-----Create Submission ...")
    submission = create_submission(result, score, DESCRIPTION, name)
    # "Optimizer1_a_example_1000.sub"
    
    print(f"-----Score: {score} in {toc-tic:.4f} s")
    
print("RESULT:")
print("--------------------------------------------------------------------------")

for name,score,timing in zip(FILENAMES, scores, times):
    print(f"{name:<30} | { score:<13} | {timing:.4f} s")

print("--------------------------------------------------------------------------")
print(f"  -> Full Score {np.sum(score):.0f}")
```
